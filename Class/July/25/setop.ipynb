```python
def group2():
    ai_students= {"Amit", "Bina", "Chetan", "Divya"}
    cyber_students= {"Chetan", "Divya", "Esha", "Farhan"}

    print("what do you wanna do?")
    print("1. Students who attended both workshops")
    print("2. Students who attended only the AI workshop")
    print("3. Students who attended at least one workshop")
    print("4. Did all AI workshop students also attend Cybersecurity?")
    choice = int(input("Enter your choice (1-4): "))

    match choice:
        case 1:
            both= ai_students&cyber_students
            print("Both workshops:", both)
            
        case 2:    
            only_ai= ai_students-cyber_students
            print("Only AI workshop:", only_ai)
        
        case 3:
            any_one= ai_students|cyber_students
            print("At least one workshop:", any_one)

        case 4:
            ai_in_cyber= ai_students.issubset(cyber_students)
            print("All AI students also in Cybersecurity?", ai_in_cyber)
```

```python
def group1():
    raj_books = {"Python Basics", "Data Structures", "Algorithms", "Machine Learning"}
    ranjith_books = {"Data Structures", "Machine Learning", "Deep Learning", "Computer Networks"}

    print("1. Books borrowed by both Raj and Ranjith")
    print("2. Books only Raj borrowed")
    print("3. Books borrowed by either Raj or Ranjith, but not both")
    print("4. Did Raj borrow only books also borrowed by Ranjith?")
    choice = int(input("Enter your choice (1-4): "))

    match choice:
        case 1:
            print("Both:", raj_books & ranjith_books)
        case 2:
            print("Only Raj:", raj_books - ranjith_books)
        case 3:
            print("Either but not both:", raj_books ^ ranjith_books)
        case 4:
            print("Raj subset of Ranjith:", raj_books.issubset(ranjith_books))
```

```python
def group3():
    books_jan = {"Python", "Java", "C++", "HTML"}
    books_feb = {"Python", "C++", "JavaScript", "CSS"}

    print("1. Books issued in both January and February")
    print("2. Books new in February")
    print("3. Books issued in either month but not both")
    print("4. Remove books from Jan that were also in Feb (update Jan set)")
    choice = int(input("Enter your choice (1-4): "))

    match choice:
        case 1:
            print("Both months:", books_jan & books_feb)
        case 2:
            print("New in Feb:", books_feb - books_jan)
        case 3:
            print("Either month but not both:", books_jan ^ books_feb)
        case 4:
            books_jan.difference_update(books_feb)
            print("Updated Jan books:", books_jan)
```

```python
def group4():
    python_team = {"John", "Lily", "Ravi"}
    java_team = {"Lily", "Ravi", "Sara"}

    print("1. Employees who know both Python and Java")
    print("2. Employees who know only Java")
    print("3. Employees who know either Python or Java, but not both")
    print("4. Is Python team a subset of Java team?")
    choice = int(input("Enter your choice (1-4): "))

    match choice:
        case 1:
            print("Both skills:", python_team & java_team)
        case 2:
            print("Only Java:", java_team - python_team)
        case 3:
            print("Either but not both:", python_team ^ java_team)
        case 4:
            print("Python ⊆ Java?", python_team.issubset(java_team))
```

```python
def group5():
    data_science = {"Anil", "Bhavya", "Chetan", "Deepa"}
    cloud_computing = {"Chetan", "Deepa", "Eshan", "Farah"}

    print("1. Students in both courses")
    print("2. Students in only one course")
    print("3. Total unique students")
    print("4. Update data_science to only students who took both")
    choice = int(input("Enter your choice (1-4): "))

    match choice:
        case 1:
            print("Both:", data_science & cloud_computing)
        case 2:
            print("Only one course:", data_science ^ cloud_computing)
        case 3:
            print("Total unique:", len(data_science | cloud_computing))
        case 4:
            data_science.intersection_update(cloud_computing)
            print("Updated data_science:", data_science)
```

```python
def group6():
    cart_user1 = {"Shoes", "Watch", "T-shirt"}
    cart_user2 = {"Watch", "Jeans", "Shoes"}

    print("1. Common items")
    print("2. Unique items to user2")
    print("3. Total unique items")
    print("4. Do both carts have the same items?")
    choice = int(input("Enter your choice (1-4): "))

    match choice:
        case 1:
            print("Common:", cart_user1 & cart_user2)
        case 2:
            print("Unique to user2:", cart_user2 - cart_user1)
        case 3:
            print("Total unique items:", cart_user1 | cart_user2)
        case 4:
            print("Same items?", cart_user1 == cart_user2)
```

```python
def main():
    while True:
        print("\n===== Set Operation Menu =====")
        print("1. Group 1")
        print("2. Group 2")
        print("3. Group 3")
        print("4. Group 4")
        print("5. Group 5")
        print("6. Group 6")
        print("7. Exit")
        choice = int(input("Select a group to run (1-7): "))

        match choice:
            case 1:
                group1()
            case 2:
                group2() 
            case 3:
                group3()
            case 4:
                group4()
            case 5:
                group5()
            case 6:
                group6()
            case 7:
                print("Exiting program.")
                break
            case _:
                print("Invalid choice. Please enter 1–7.")
                
if __name__ == "__main__":
    main()
```
